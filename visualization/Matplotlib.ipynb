```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2
x, y
```

# Functional method

```python
''' Function '''
plt.plot(x, y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Title')
# plt.show()
```

```python
''' Multi plot in one'''
plt.subplot(1,2,1)
plt.plot(x, y, 'r')

plt.subplot(1,2,2)
plt.plot(y, x, 'b')
```

# Object oriented method

```python
''' Object oriented method '''
fig = plt.figure()

axes  = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x, y)
axes.set_xlabel('x-axis')
axes.set_ylabel('y-axis')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes_0 = fig.add_axes([0,0,1,1])  # [shift to left, shift to bottom, plot width, plot height]
axes_1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes_2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes_1.plot(x, y)
axes_1.set_title('Plot 1')

axes_2.plot(y, x)
axes_2.set_title('Plot 2')
```

```python
''' subplots '''
fig, axes = plt.subplots(nrows=1, ncols=2)
plt.tight_layout()
# for ax in axes:
#     ax.plot(x,y)
axes[0].plot(x,y)
axes[0].set_title('Plot 0')

axes[1].plot(y,x)
axes[1].set_title('Plot 1')
```

## Figure size and DPI

```python
fig = plt.figure(figsize=(4,2), dpi=100)

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(8,4))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

## Legend

```python
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(4,4))

axes.plot(x,x**2, label = 'x-2')
axes.plot(x,x**3, label = 'x-3')

axes.legend(loc=0) # loc=0: best
```

## Plot appearance

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,
        color='blue',
        linewidth=2,
        alpha=0.5,
        linestyle='--',
        marker='o',
        markersize=20,
        markerfacecolor='yellow',
        markeredgewidth=3)
### color: blue, orange, green, purple, yellow; #FF8c00
### linewidth (times of default): 1, 2, 3
### alpha (transparency)
### linestyle: '-', '--', '-.', ':', 'steps'
### marker: 'o', '+', '*', '1'
### markersize: 1, 5, 10, 20
```

## Axis customize

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,
        color='purple',
        lw=2,
        ls='--')
ax.set_xlim([0,3])
ax.set_ylim([0,5])
```
